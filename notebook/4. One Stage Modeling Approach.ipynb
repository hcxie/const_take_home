# Smooth the prediction

```python
def smooth(y, box_pts):
    box = np.ones(box_pts)/box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth

pred_df = pd.read_csv(input_path + "stack_one_stage_submission2.csv")
y = pred_df["total_cases"]
y1 = smooth(y, 4)
plt.plot(y)
plt.plot(y1)
pred_df["total_cases"] = y1.astype(int)
pred_df.to_csv("stack_one_stage_submission2.csv", index = False)
```
