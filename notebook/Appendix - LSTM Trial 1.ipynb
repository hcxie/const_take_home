# LSTM Trial 1

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import torch
import torch.nn as nn
from torch.autograd import Variable
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from scipy.stats.mstats import winsorize
```

```python
# load data
train_data = pd.read_csv("data/dengue_features_train.csv")
train_label = pd.read_csv("data/dengue_labels_train.csv")
test_data = pd.read_csv("data/dengue_features_test.csv")
```

```python
# LSTM class
class EarlyStopping():
    """
    Early stopping to stop the training when the loss does not improve after
    certain epochs.
    """
    def __init__(self, patience=5, min_delta=0):
        """
        param patience: how many epochs to wait before stopping when loss is
            not improving
        param min_delta: minimum difference between new loss and old loss for
        new loss to be considered as an improvement
        """
        self.patience = patience
        self.min_delta = min_delta
        self.counter = 0
        self.best_loss = None
        self.early_stop = False
        pass
    
    def __call__(self, val_loss):
        if self.best_loss == None:
            self.best_loss = val_loss
        elif self.best_loss - val_loss > self.min_delta:
            self.best_loss = val_loss
        # reset counter if validation loss improves
            self.counter = 0
        elif self.best_loss - val_loss < self.min_delta:
            self.counter += 1
            print(f"INFO: Early stopping counter {self.counter} of {self.patience}")
            
            if self.counter >= self.patience:
                print('INFO: Early stopping')
                self.early_stop = True
        return

class LSTM(nn.Module):
    def __init__(self, input_size=1, hidden_layer_size=100, output_size=1):
        super().__init__()
        self.hidden_layer_size = hidden_layer_size

        self.lstm = nn.LSTM(input_size, hidden_layer_size)

        self.linear = nn.Linear(hidden_layer_size, output_size)

        self.hidden_cell = (torch.zeros(1,1,self.hidden_layer_size),
                            torch.zeros(1,1,self.hidden_layer_size))

    def forward(self, input_seq):
        lstm_out, self.hidden_cell = self.lstm(input_seq.view(len(input_seq) ,1, -1), self.hidden_cell)
        predictions = self.linear(lstm_out.view(len(input_seq), -1))
        return predictions[-1]
```

```python
class LSTM_wrapper():
    def __init__(self,
                 model,
                 early_stopper,
                 criterion, #torch.nn.MSELoss()
                 eta = 0.0001,                
                 time_window = 52,
                 upper_threshold = 90):
        
        self.model = model
        
        self.criterion = criterion
        
        self.eta = eta
        
        self.optimizer = None
        
        self.tw = time_window
        
        self.scaler = None
        
        self.early_stopping = early_stopper
        
        self.upper_threshold = upper_threshold
             
        pass
    
    # data preparation functions
    def _LSTM_sliding_windows(self, input_data):
        inout_seq = []
        L = len(input_data)
        for i in range(L-self.tw):
            train_seq = input_data[i:i+self.tw]
            train_label = input_data[i+self.tw:i+self.tw+1]
            inout_seq.append((train_seq ,train_label))
        return inout_seq

    def _LSTM_create_data(self, train_X):
        self.scaler = StandardScaler()
        train_data_normalized = self.scaler.fit_transform(train_X.reshape(-1, 1))
        train_data_normalized = torch.FloatTensor(train_data_normalized).view(-1)
        full_data = self._LSTM_sliding_windows(train_data_normalized)
        train_data = full_data[0:int(len(full_data)*0.85)]
        val_data = full_data[int(len(full_data)*0.85):]    
        return full_data, train_data, val_data
    

    def _LSTM_train_eval(self, epoch, train_data, val_data):
        self.model.train()
    
        train_loss = 0
        for x, y in train_data:
            self.optimizer.zero_grad()
            self.model.hidden_cell = (torch.zeros(1, 1, self.model.hidden_layer_size),
                                     torch.zeros(1, 1, self.model.hidden_layer_size))
            outputs = self.model(x)
            loss = self.criterion(outputs, y)
            train_loss += loss
            loss.backward()
            self.optimizer.step()
            
        
        train_loss /= len(train_data)
        
        self.model.eval()
    
        val_loss = 0
        with torch.no_grad():
            for x_val, y_val in val_data:
                val_outputs = self.model(x_val)
                loss = self.criterion(val_outputs, y_val)
                val_loss += loss
            
        val_loss /= len(val_data)
        
        
        if epoch % 10 == 0:
            print('Epoch: {}. train Loss: {}. val loss: {}'.format(epoch, train_loss, val_loss))
        
        return train_loss, val_loss
        

    def _LSTM_predict_seq(self, input_data, pred_len):
    
        pred_input_normalized = self.scaler.transform(input_data.reshape(-1, 1))
        pred_input_normalized = torch.FloatTensor(pred_input_normalized).view(-1)
        
        pred_input = pred_input_normalized[-self.tw:].tolist()
        
        self.model.eval()
        for i in range(pred_len):
            seq = torch.FloatTensor(pred_input[-self.tw:])
            with torch.no_grad():
                self.model.hidden = (torch.zeros(1, 1, self.model.hidden_layer_size),
                                    torch.zeros(1, 1, self.model.hidden_layer_size))
                pred_input.append(self.model(seq).item())
                        
        actual_pred = self.scaler.inverse_transform(np.array(pred_input[self.tw:] ).reshape(-1, 1))
        #actual_pred = pred_input[self.tw:].reshape(-1, 1)
            
        return actual_pred

    def plot_one_prediction(self, predict, title):
        data_predict = predict
        plt.plot(data_predict)
        plt.suptitle(title)
        plt.show()
        return
    
    def plot_two_prediction(self,
                            predict1,
                            predict2, 
                            title):
        data_predict1 = predict1
        data_predict2 = predict2
        plt.plot(data_predict1)
        plt.plot(data_predict2)
        plt.suptitle(title)
        plt.show()
        return
    
    def smooth_data(self,
                    train_data):
        one_minus_upper_bound = 1-(self.upper_threshold/100)
        smooth_data = winsorize(train_data, limits=[0, one_minus_upper_bound], inclusive=(True, False)).data
        
        return smooth_data


    def LSTM_main(self, city, train_data, test_data, nums_epochs):
        
        train_X = train_data.loc[train_data["city"] == city,:].copy()
        train_X = train_X["total_cases"].values.astype(float)
        
        train_X = self.smooth_data(train_X)
           
        test_data = test_data.loc[test_data["city"] == city,:].copy()
    
        full_data, train_data, val_data= self._LSTM_create_data(train_X)
    
        self.optimizer = torch.optim.Adam(self.model.parameters(), lr=self.eta)
    
        # Train the model
        for epoch in range(nums_epochs):
            print("current epoch: ", epoch, end = "\r", flush = True)
            train_loss, val_loss = self._LSTM_train_eval(epoch, 
                                                         train_data, 
                                                         val_data)
            self.early_stopping(val_loss)
         
            if self.early_stopping.early_stop:
                print('Early stopped at Epoch: {}. train Loss: {}. val loss: {}'.format(epoch, train_loss, val_loss))
                break
            
    
        # plot the prediction on training data
        train_predict = self._LSTM_predict_seq(train_X[0:self.tw], len(train_X) - self.tw)
        #self.plot_two_prediction(train_predict, train_X[self.tw:])
        self.plot_one_prediction(train_predict, "{} train data prediction".format(city))
        self.plot_one_prediction(train_X[self.tw:], "{} train data actual".format(city))
        
        test_predict = self._LSTM_predict_seq(train_X[len(train_X) - self.tw:], len(test_data))
    
        self.plot_one_prediction(test_predict, "{} test data prediction".format(city))
    
        return test_predict
```

```python

time_window = 52

model_sj = LSTM()
criterion_sj = torch.nn.MSELoss()
early_stopper_sj = EarlyStopping(5, 0)
main_sj = LSTM_wrapper(model = model_sj, 
                    criterion = criterion_sj, 
                    early_stopper = early_stopper_sj,
                    time_window = time_window,
                    upper_threshold = 60)

model_iq = LSTM()
criterion_iq = torch.nn.MSELoss()
early_stopper_iq = EarlyStopping(5, 0)
main_iq = LSTM_wrapper(model = model_iq, 
                      criterion = criterion_iq, 
                      early_stopper = early_stopper_iq,
                      time_window = time_window,
                      upper_threshold = 30)
```

```python
predict_sj = main_sj.LSTM_main("sj", train_label, test_data, 2000)
predict_iq = main_iq.LSTM_main("iq", train_label, test_data, 2000)

predict_sj = predict_sj.reshape(-1)
predict_iq = predict_iq.reshape(-1)

prediction = np.concatenate([predict_sj,predict_iq])

submission = pd.read_csv("data/submission_format.csv")
submission["total_cases"] = prediction

submission.to_csv("lstm_prediction.csv", index = False)
```
