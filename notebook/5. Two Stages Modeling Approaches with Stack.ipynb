# Combine the prediction from one-stage model and two-stage model

```python
def smooth(y, box_pts):
    box = np.ones(box_pts)/box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth
```

```python
# stacking perpensity-severity to one stage prediction
submission1 = pd.read_csv("data/clf_stack_pred_2.csv")
submission2 = pd.read_csv("data/sev_stack_pred2.csv")
submission3 = pd.read_csv("data/stack_one_stage_submission2.csv")
```

```python
test_pred_rev_mask = (submission1["predict_basedon_prob"]>0).astype(int)
test_pred_rev_mask = 1 - test_pred_rev_mask
y = (test_pred_rev_mask*submission3["total_cases"] 
                              + (submission1["predict_basedon_prob"]>0) 
                              * submission2["total_cases"])
y = y.astype(int)
```

```python
y1 = smooth(y, 2)

plt.plot(y, label = "two stages model")
plt.plot(submission3["total_cases"], label = "one stage model")
plt.legend()
plt.xlabel("time")
plt.ylabel("cases")
plt.title("two stages model vs one stage model")
```

```python
submission3["total_cases"] =y1.astype(int)
submission3.to_csv("stack_two_stage_submission2.csv", index = False)
```
